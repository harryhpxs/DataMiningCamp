```python
import pandas
pandas.set_option('display.max_columns', 100)
df = pandas.read_csv('data.csv', encoding = 'GBK')
```

```python
# 查看是否有重复值
df['id_name'].duplicated().value_counts()
```

通过以上信息可观察到，部分无关特征可被删除：

Unnamed: 0, custid, trade_no, bank_card_no, id_name
    
其中：
- Unamed: 0 为递增数字，疑似原数据中对应index
- custid 客户ID，于该数据集中无重复
- trade_no 交易号，同上
- bank_card_no 脱敏后均为"卡号1"
- id_name 客户姓名，有重复但应为同名的不同客户

```python
df = df.drop(['Unnamed: 0', 'custid', 'trade_no', 'bank_card_no', 'id_name'],axis = 1)
```

```python
# 补充缺失值
df = df.fillna(df.mean)
```

```python
# 转换日期类型
df['latest_query_time'] = pandas.to_datetime(df['latest_query_time'])
df['loans_latest_time'] = pandas.to_datetime(df['loans_latest_time'])
```

```python
# 切分数据集 测试集30%，训练集70%
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.3, random_state=2018)
```
